```python
import pandas as pd

df = pd.ExcelFile('Data_steward_excel_test.xlsx')
price = pd.read_excel(df, 'Цены')
discount = pd.read_excel(df, 'Скидка')
discount = discount.rename(columns = {'Region' : 'Channel'})
data = pd.DataFrame({'SKU': ['Product 1', 'Product 2', 'Product 3', 'Product 4', 'Product 5',
                            'Product 6', 'Product 7', 'Product 8', 'Product 9', 'Product 10']})
data = data.set_index('SKU')
```

```python
for i in range(2001,2011,2):
    d = pd.read_excel(df, str(i)).merge(discount, on='Channel')
    d = d[d['Channel'] == 'Commercial']
    d = d.merge(price[price['Year'] == i], on ='SKU')
    total_com = sum(d['SO Packs']*d['Price'])
    d = d[d['Discount?'] == '+']
    d['sum_disc'] = d['SO Packs']*d['Discount']*d['Price']
    sumd = d.groupby(d['SKU'])['sum_disc'].sum()
    data['sum'+str(i)] = round(sumd)
    proportion = round(sumd/total_com*100,4)
    data['%'+str(i)] = proportion
```

```python
data.to_excel('data.xlsx')
```

```python
df = pd.ExcelFile('Data_steward_excel_test.xlsx')
price = pd.read_excel(df, 'Цены')
discount = pd.read_excel(df, 'Скидка')
discount = discount.rename(columns = {'Region' : 'Channel'})
data2 = pd.DataFrame()
```

```python
for i in range(2001,2010):
    d = pd.read_excel(df, str(i)).merge(discount, on='Channel')
    d = d.merge(price[price['Year'] == i], on ='SKU')
    d['Sum_disc'] = d['SO Packs']*d['Discount']*d['Price']
    d.loc[d['Discount?']!='+', 'Sum_disc'] = 0
    d['Sum'] = d['SO Packs']*d['Price']
    data2 = pd.concat([data2,d], axis=0)
    
```
